## 4 - Density estimation using Parzen densities **(Coding)**

Next to classifiers, PRTools also has the possibility to estimate densities. In this section we are going to estimate the density using a Parzen density estimator, called parzenm in PRTools

````python
#We start with creating a simple dataset with:
a = pr.gendats([20,20],1,8)

#We define the width parameter h for the Parzen kernel:
h = 0.5

#The function parzenm estimates a density for a given dataset. In most cases a PRTools prdataset is labeled, and these labels are used in the function parzenm to estimate a density for each class. To define a Parzen density estimator with a certain width parameter h on the entire dataset, ignoring labels, type:
a = pr.prdataset(+a)
w = pr.parzenm(a,h)

#This mapping can now be plotted along with the data:
pr.scatterd(a)
pr.plotm(w)

#If your graphs look a little “bumpy”, you can increase the grid size PRTools uses for plotting:

pr.plotm(w,gridsize=100)
````

When you want to evaluate the fit of a density model to some data, you have to define an error. One possibility is to use the log-likelihood, defined as:

$$
LL(X) = \log \prod_i \hat{p}(x_i) = \sum_i \log(\hat{p}(x_i)) \tag{1.2}
$$

The better the data $x$ fits in the probability density model $\hat{p}$, the higher the values of $\hat{p}(x)$ will be. This will result in a high value of $\sum_i \log(\hat{p}(x_i))$. When we have different probability density estimates $\hat{p}$, we use the one which has the highest value of $LL$.

Note that when we fill in different values for the width parameters $h$ in the Parzen density estimation, we have different estimates $\hat{p}$. Using the log-likelihood as a criterion, we can optimize the value of this free parameter $h$ to maximize $LL$.

To get an honest estimate of the log-likelihood, we have to evaluate the log-likelihood (1.2) on a test set. That means that we have to make (or measure) new data from the same distribution as where the training data came from. When we evaluate the log-likelihood on the data on which the probability density was fitted, we would get a too optimistic estimation of the error. We might conclude that we have fitted the data very well, while actually a new dataset from the same distribution does not fit in the density at all! Therefore, if you want to evaluate the performance of an estimated $\hat{p}$, use an independent test set!


### **Exercise 4.1**

Use the data from the same distribution as in the previous exercise to train a Parzen density estimator for different values of h. Compute the log-likelihood of this training set given the estimated densities (for different $h$)

````python
a = pr.gendats([20,20],1,8)                 # Generate data
a = pr.prdataset(+a)
hs = [0.01,0.05,0.1,0.25,0.5,1,1.5,2,3,4,5] # Array of h’s to try
LL = np.zeros(len(hs))
for i in range(len(hs)):
                                         # For each h...
    w = pr.parzenm(a,hs[i])                  #   estimate Parzen density
    LL[i] = np.sum(np.log(+(a*w)));          #   calculate log-likelihood
plt.plot(hs,LL)
plt.xlabel('h')
plt.ylabel('log-likelihood')
plt.title('Parzen density estimation')
plt.show()
````

## Understanding `pr.gendats([10, 10])` - Different Means, Equal Covariance

Let's explore what `pr.gendats([10, 10])` actually generates and verify that it creates two Gaussian classes with different means but equal covariance matrices.


````python
# Generate the dataset
d = pr.gendats([10, 10])
print("Dataset info:", d)
print()

# Extract the data and labels
data = +d  # Convert to numpy array
labels = d.labels  # Get the labels

# Separate the two classes
class1_data = data[labels.flatten() == 0]  # Class 1 (label 0)
class2_data = data[labels.flatten() == 1]  # Class 2 (label 1)

print("Class 1 data shape:", class1_data.shape)
print("Class 2 data shape:", class2_data.shape)
print()

# Calculate means for each class
mean1 = np.mean(class1_data, axis=0)
mean2 = np.mean(class2_data, axis=0)

print("Class 1 mean:", mean1)
print("Class 2 mean:", mean2)
print("Difference in means:", mean2 - mean1)
print()

# Calculate covariance matrices for each class
cov1 = np.cov(class1_data.T)
cov2 = np.cov(class2_data.T)

print("Class 1 covariance matrix:")
print(cov1)
print()
print("Class 2 covariance matrix:")
print(cov2)
print()
print("Difference in covariance matrices:")
print(cov2 - cov1)
print()

# Visualize the dataset
pr.scatterd(d)
plt.title('Two Gaussian Classes: Different Means, Equal Covariance')
plt.grid(True, alpha=0.3)
plt.show()

print("CONCLUSION:")
print("✓ The classes have DIFFERENT means")
print("✓ The classes have approximately EQUAL covariance matrices (small differences due to sampling)")
print("✓ This confirms that pr.gendats([10, 10]) generates two Gaussian classes with different means but equal covariance matrices")
````

### PRTools Key Concepts and Functions

**Dataset Generation:**
- `pr.gendats([n1, n2])` - Two Gaussian classes with different means, equal covariance
- `pr.gendatb([n1, n2])` - Banana-shaped dataset  
- `pr.gendatc([n1, n2])` - Circular classes
- `pr.gendatd([n1, n2])` - Difficult/complex dataset

**Classifiers:**
- `pr.ldc()` - Linear Discriminant Classifier (assumes equal covariance)
- `pr.qdc()` - Quadratic Discriminant Classifier (allows different covariance)
- `pr.svc()` - Support Vector Classifier
- `pr.knnc(k)` - k-Nearest Neighbor Classifier
- `pr.nmc()` - Nearest Mean Classifier

**Workflow:**
1. **Generate/Load Data**: `data = pr.gendats([50, 50])`
2. **Create Classifier**: `classifier = pr.ldc()`
3. **Train**: `classifier.train(data)`
4. **Evaluate**: `predictions = classifier.eval(data)`
5. **Get Error**: `error = pr.testc(predictions)`
6. **Visualize**: `pr.scatterd(data); pr.plotc(classifier)`


### Why This Matters for Classification

When you have **different means but equal covariance matrices**, this is the ideal scenario for **Linear Discriminant Analysis (LDA)**:

1. **Linear Decision Boundary**: The optimal Bayes decision boundary will be linear
2. **LDC vs QDC**: Linear Discriminant Classifier (`pr.ldc()`) should perform as well as Quadratic Discriminant Classifier (`pr.qdc()`)
3. **Theoretical Foundation**: This matches the assumptions of LDA perfectly

Let's demonstrate this:


````python
# Generate a larger dataset for better demonstration
d = pr.gendats([100, 100])

# Train both Linear and Quadratic Discriminant Classifiers
ldc_classifier = pr.ldc()
qdc_classifier = pr.qdc()

ldc_classifier.train(d)
qdc_classifier.train(d)

# Evaluate both classifiers
ldc_predictions = ldc_classifier.eval(d)
qdc_predictions = qdc_classifier.eval(d)

ldc_error = pr.testc(ldc_predictions)
qdc_error = pr.testc(qdc_predictions)

print(f"Linear Discriminant Classifier Error: {ldc_error:.4f}")
print(f"Quadratic Discriminant Classifier Error: {qdc_error:.4f}")
print()

# Visualize both decision boundaries
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
pr.scatterd(d)
plt.title('Original Data\n(Different Means, Equal Covariance)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
pr.scatterd(d)
pr.plotc(ldc_classifier)
plt.title(f'Linear Discriminant Classifier\nError: {ldc_error:.4f}')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
pr.scatterd(d)
pr.plotc(qdc_classifier)
plt.title(f'Quadratic Discriminant Classifier\nError: {qdc_error:.4f}')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("OBSERVATION:")
print("✓ Both classifiers perform similarly because the data satisfies LDA assumptions")
print("✓ The decision boundaries are nearly identical")
print("✓ This confirms that pr.gendats() creates the ideal scenario for linear classification")
````

## Creating Gaussian Classes with Different Covariance Matrices

Now let's explore how to create two Gaussian classes with **different covariance matrices**. This is the scenario where Quadratic Discriminant Analysis (QDA) shines!


### Method 1: Using PRTools Built-in Functions

Some PRTools dataset generators create classes with different covariance matrices by default:


````python
# Method 1a: gendatd - "Difficult" dataset with different covariances
print("=== Method 1a: pr.gendatd() - Difficult Dataset ===")
d_difficult = pr.gendatd([50, 50])

# Extract data and analyze covariance matrices
data_diff = +d_difficult
labels_diff = d_difficult.labels

class1_diff = data_diff[labels_diff.flatten() == 0]
class2_diff = data_diff[labels_diff.flatten() == 1]

cov1_diff = np.cov(class1_diff.T)
cov2_diff = np.cov(class2_diff.T)

print("Class 1 mean:", np.mean(class1_diff, axis=0))
print("Class 2 mean:", np.mean(class2_diff, axis=0))
print()
print("Class 1 covariance:")
print(cov1_diff)
print()
print("Class 2 covariance:")
print(cov2_diff)
print()
print("Covariance difference:")
print(cov2_diff - cov1_diff)

# Visualize
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
pr.scatterd(d_difficult)
plt.title('gendatd(): Different Covariances')
plt.grid(True, alpha=0.3)

# Train LDC and QDC to see the difference
ldc_diff = pr.ldc()
qdc_diff = pr.qdc()
ldc_diff.train(d_difficult)
qdc_diff.train(d_difficult)

ldc_error_diff = pr.testc(ldc_diff.eval(d_difficult))
qdc_error_diff = pr.testc(qdc_diff.eval(d_difficult))

plt.subplot(1, 3, 2)
pr.scatterd(d_difficult)
pr.plotc(ldc_diff)
plt.title(f'LDC Error: {ldc_error_diff:.4f}')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
pr.scatterd(d_difficult)
pr.plotc(qdc_diff)
plt.title(f'QDC Error: {qdc_error_diff:.4f}')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"LDC Error: {ldc_error_diff:.4f} (assumes equal covariance)")
print(f"QDC Error: {qdc_error_diff:.4f} (allows different covariance)")
print("→ QDC should perform better when covariances are truly different!")
````

### Method 2: Custom Generation with NumPy

Create your own Gaussian classes with completely custom means and covariance matrices:


````python
print("=== Method 2: Custom Gaussian Classes with Different Covariances ===")

# Define parameters for two classes
n_samples = 100

# Class 1: Circular (isotropic) covariance
mean1 = np.array([0, 0])
cov1 = np.array([[1.0, 0.0],    # Circular: equal variance, no correlation
                 [0.0, 1.0]])

# Class 2: Elongated covariance with correlation
mean2 = np.array([3, 2])
cov2 = np.array([[2.0, 1.5],    # Elongated: different variances + correlation
                 [1.5, 3.0]])

# Generate samples using numpy's multivariate_normal
np.random.seed(42)  # For reproducibility
class1_samples = np.random.multivariate_normal(mean1, cov1, n_samples)
class2_samples = np.random.multivariate_normal(mean2, cov2, n_samples)

# Combine into single dataset
X_custom = np.vstack([class1_samples, class2_samples])
y_custom = np.hstack([np.zeros(n_samples), np.ones(n_samples)]).reshape(-1, 1)

# Create PRTools dataset
d_custom = pr.prdataset(X_custom, y_custom)

print("Custom dataset info:", d_custom)
print()
print("Designed Class 1 mean:", mean1)
print("Designed Class 1 covariance:")
print(cov1)
print()
print("Designed Class 2 mean:", mean2) 
print("Designed Class 2 covariance:")
print(cov2)
print()

# Verify actual sample statistics
actual_cov1 = np.cov(class1_samples.T)
actual_cov2 = np.cov(class2_samples.T)

print("Actual Class 1 mean:", np.mean(class1_samples, axis=0))
print("Actual Class 1 covariance:")
print(actual_cov1)
print()
print("Actual Class 2 mean:", np.mean(class2_samples, axis=0))
print("Actual Class 2 covariance:")
print(actual_cov2)

# Visualize with confidence ellipses
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
pr.scatterd(d_custom)
plt.title('Custom Classes:\nDifferent Means & Covariances')
plt.grid(True, alpha=0.3)

# Add confidence ellipses to show covariance structure
from matplotlib.patches import Ellipse
def plot_confidence_ellipse(mean, cov, ax, n_std=2.0, **kwargs):
    """Plot confidence ellipse for 2D Gaussian"""
    eigenvals, eigenvecs = np.linalg.eigh(cov)
    angle = np.degrees(np.arctan2(eigenvecs[1, 0], eigenvecs[0, 0]))
    width, height = 2 * n_std * np.sqrt(eigenvals)
    ellipse = Ellipse(mean, width, height, angle=angle, fill=False, **kwargs)
    ax.add_patch(ellipse)

ax = plt.gca()
plot_confidence_ellipse(mean1, cov1, ax, color='blue', linewidth=2, linestyle='--')
plot_confidence_ellipse(mean2, cov2, ax, color='red', linewidth=2, linestyle='--')

# Train classifiers
ldc_custom = pr.ldc()
qdc_custom = pr.qdc()
ldc_custom.train(d_custom)
qdc_custom.train(d_custom)

ldc_error_custom = pr.testc(ldc_custom.eval(d_custom))
qdc_error_custom = pr.testc(qdc_custom.eval(d_custom))

plt.subplot(1, 3, 2)
pr.scatterd(d_custom)
pr.plotc(ldc_custom)
plt.title(f'LDC Error: {ldc_error_custom:.4f}\n(Wrong assumption!)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
pr.scatterd(d_custom)
pr.plotc(qdc_custom)
plt.title(f'QDC Error: {qdc_error_custom:.4f}\n(Correct assumption!)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"LDC Error: {ldc_error_custom:.4f} (assumes equal covariance - WRONG!)")
print(f"QDC Error: {qdc_error_custom:.4f} (allows different covariance - CORRECT!)")
print("→ QDC performs much better because it matches the true data distribution!")
````

### Method 3: Extreme Covariance Differences

Let's create an extreme example to really see the difference:


````python
print("=== Method 3: Extreme Covariance Differences ===")

# Class 1: Very tight, circular distribution
mean1_extreme = np.array([1, 1])
cov1_extreme = np.array([[0.1, 0.0],    # Very small variance
                         [0.0, 0.1]])

# Class 2: Very elongated distribution
mean2_extreme = np.array([3, 3])
cov2_extreme = np.array([[4.0, 0.0],    # High variance in x, low in y
                         [0.0, 0.1]])

# Generate samples
np.random.seed(123)
class1_extreme = np.random.multivariate_normal(mean1_extreme, cov1_extreme, 100)
class2_extreme = np.random.multivariate_normal(mean2_extreme, cov2_extreme, 100)

# Create dataset
X_extreme = np.vstack([class1_extreme, class2_extreme])
y_extreme = np.hstack([np.zeros(100), np.ones(100)]).reshape(-1, 1)
d_extreme = pr.prdataset(X_extreme, y_extreme)

print("Extreme Class 1 covariance (tight circle):")
print(cov1_extreme)
print()
print("Extreme Class 2 covariance (elongated):")
print(cov2_extreme)
print()

# Train classifiers
ldc_extreme = pr.ldc()
qdc_extreme = pr.qdc()
ldc_extreme.train(d_extreme)
qdc_extreme.train(d_extreme)

ldc_error_extreme = pr.testc(ldc_extreme.eval(d_extreme))
qdc_error_extreme = pr.testc(qdc_extreme.eval(d_extreme))

# Visualize
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
pr.scatterd(d_extreme)
plt.title('Extreme Covariance Differences:\nTight Circle vs Elongated')
plt.grid(True, alpha=0.3)
plt.axis('equal')

plt.subplot(1, 3, 2)
pr.scatterd(d_extreme)
pr.plotc(ldc_extreme)
plt.title(f'LDC Error: {ldc_error_extreme:.4f}\n(Terrible fit!)')
plt.grid(True, alpha=0.3)
plt.axis('equal')

plt.subplot(1, 3, 3)
pr.scatterd(d_extreme)
pr.plotc(qdc_extreme)
plt.title(f'QDC Error: {qdc_error_extreme:.4f}\n(Much better!)')
plt.grid(True, alpha=0.3)
plt.axis('equal')

plt.tight_layout()
plt.show()

print(f"LDC Error: {ldc_error_extreme:.4f}")
print(f"QDC Error: {qdc_error_extreme:.4f}")
print(f"Improvement: {((ldc_error_extreme - qdc_error_extreme) / ldc_error_extreme * 100):.1f}% reduction in error!")
````

### Summary: Creating Different Covariance Matrices

**Method 1: PRTools Built-in**
- `pr.gendatd()` - Creates "difficult" dataset with different covariances
- `pr.gendatc()` - Circular classes (can have different scales)

**Method 2: Custom with NumPy**
```python
# Define custom covariance matrices
cov1 = np.array([[var1_x, corr], [corr, var1_y]])
cov2 = np.array([[var2_x, corr], [corr, var2_y]])

# Generate samples
class1 = np.random.multivariate_normal(mean1, cov1, n_samples)
class2 = np.random.multivariate_normal(mean2, cov2, n_samples)
```

**Key Insights:**
1. **Different covariances** → Use **QDC** (Quadratic Discriminant Classifier)
2. **Equal covariances** → Use **LDC** (Linear Discriminant Classifier)
3. **Wrong assumption** → Poor performance!
4. **Covariance shapes**: Circular, elongated, rotated, correlated

**When to Use Each:**
- **LDC**: Classes have same "shape" (equal covariance)
- **QDC**: Classes have different "shapes" (different covariance)
- **SVC**: Non-linear boundaries, robust to assumptions
